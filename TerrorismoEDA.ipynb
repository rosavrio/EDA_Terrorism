```python
# imporrttamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot, plot
```

```python
# leemos el dataset que vamos a analizar
dataset = pd.read_csv('EDA_Terrorism.csv', low_memory=False, encoding="ISO-8859-1")
raw_dataset = pd.DataFrame(dataset)
raw_dataset.head()
```

```python
# comprobamos los nombres de las columnas y cuantas hay para el dataset sin tratar
print(raw_dataset.columns)
print(len(raw_dataset.columns))
```

```python
# cambiamos los nombres de las columnas a nombres mas fÃ¡ciles de tratar
raw_dataset = raw_dataset.rename(columns={'eventid': 'id',
                                          'iyear': 'year',
                                          'imonth': 'month',
                                          'iday': 'day',
                                          'extended':'extended_date',
                                          'country': 'country_code',
                                          'country_txt': 'country',
                                          'region':'region_code',
                                          'region_txt': 'region',
                                          'provstate': 'state',
                                          'attacktype1': 'attack_type_code',
                                          'attacktype1_txt': 'attack_type',
                                          'targtype1': 'target_type_code',
                                          'targtype1_txt': 'target_type',
                                          'targsubtype1': 'targ_subtype_code',
                                          'targsubtype1_txt': 'targ_subtype',
                                          'natlty1': 'targ_natlty_code',
                                          'natlty1_txt': 'targ_natlty',
                                          'gname': 'gang',
                                          'weaptype1': 'weap_type_code',
                                          'weaptype1_txt': 'weap_type',
                                          'weapsubtype1': 'weap_subtype_code',
                                          'weapsubtype1_txt': 'weap_subtype',
                                          'nkill': 'n_deaths',
                                          'nwound': 'n_wounded'})

raw_dataset.head()
```

```python
# creamos un dataset nuevo solo con las columnas que nos interesen para el tratamiento de los datos
clean_dataset = raw_dataset[['year', 'month', 'day', 'country_code', 'country', 'region_code', 'region', 'state',
                             'city', 'latitude', 'longitude', 'success', 'suicide', 'attack_type_code', 'attack_type', 'target_type_code',
                             'target_type', 'targ_subtype_code', 'targ_subtype', 'targ_natlty_code', 'targ_natlty', 'gang',
                             'weap_type_code', 'weap_type', 'weap_subtype_code', 'weap_subtype', 'n_deaths', 'n_wounded']]
clean_dataset.head()
```

```python
# echamos un vistazo a los tipos de datos con los que trabajamos
clean_dataset.dtypes
```

```python
# rellenamos valores missing con 0 y 'unknown'
clean_dataset['targ_subtype_code'] = clean_dataset['targ_subtype_code'].fillna(0)
clean_dataset['targ_subtype'] = clean_dataset['targ_subtype'].fillna('Unknown')
clean_dataset['targ_natlty_code'] = clean_dataset['targ_natlty_code'].fillna(0)
clean_dataset['targ_natlty'] = clean_dataset['targ_natlty'].fillna('Unknown')
clean_dataset['gang'] = clean_dataset['gang'].fillna('Unknown')
clean_dataset['weap_subtype_code'] = clean_dataset['weap_subtype_code'].fillna(0)
clean_dataset['weap_subtype'] = clean_dataset['weap_subtype'].fillna('Unknown')
clean_dataset['city'] = clean_dataset['city'].fillna('Unknown')
clean_dataset['n_deaths'] = clean_dataset['n_deaths'].fillna(0)
clean_dataset['n_wounded'] = clean_dataset['n_wounded'].fillna(0)

#clean_dataset.info()
```

```python
# aqui cambiamos los floats a integers para los valores numericos que no tengan sentido con decimales
clean_dataset['targ_subtype_code'] = clean_dataset['targ_subtype_code'].astype(int)
clean_dataset['targ_natlty_code'] = clean_dataset['targ_natlty_code'].astype(int)
clean_dataset['weap_subtype_code'] = clean_dataset['weap_subtype_code'].astype(int)
clean_dataset['n_deaths'] = clean_dataset['n_deaths'].astype(int)
clean_dataset['n_wounded'] = clean_dataset['n_wounded'].astype(int)
```

```python
# tiramos los registros duplicados
clean_dataset.drop_duplicates(inplace=True)
clean_dataset.head()
```

```python
'''en este paso refinamos la comprobacion de los duplicados con una combinacion de variables concretas. 
Este paso es necesario porque en ocasiones se han registrado los mismos atentados mas de una vez
con una diferencia en solo una o dos columnas, lo que hace que se contabilicen (en el caso de haberlas) mas victimas
o mas atentados ejecutados por los mismos atacantes, asi como un recuento mayor en ataques por ciudad, pais o region.'''

clean_dataset[clean_dataset.duplicated(subset=['year','day','month','latitude','longitude', 'gang','target_type', 'region', 'country', 'targ_natlty', 'n_deaths', 'n_wounded'])]
```

```python
# en este paso eliminamos los duplicados devueltos por el codigo de la celda anterior (2335 registros)
clean_dataset.drop_duplicates(subset=['year','day','month','latitude','longitude', 'gang','target_type', 'region', 'country', 'targ_natlty', 'n_deaths', 'n_wounded'], keep='first', inplace=True)
clean_dataset.head()
```

```python
# eliminamos registros donde el mes *Y* el dia sean 0 (desconocido)
clean_dataset = clean_dataset[(clean_dataset['day']==0) & (clean_dataset['month'] ==0) == False]
```

```python
# aqui volvemos a eliminar duplicados en los que solo el numero de victimas variaba
clean_dataset.drop_duplicates(subset=['year', 'day', 'month', 'latitude', 'longitude', 'country', 'city', 'gang', 'target_type', 'attack_type', 'weap_type'], keep='first', inplace=True)
clean_dataset.head()
```

```python
# guardamos la ultima version con los ultimos cambios en un documento .csv para el facil acceso/copia de seguridad
clean_dataset.to_csv('CleanTerrorismoEda.csv')
```
